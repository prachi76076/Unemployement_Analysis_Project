```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel(r'C:\Users\91969\OneDrive\Desktop\unemployment_data.xlsx')
print("First five rows of the dataset:")
print(data.head())
data['Date'] = pd.to_datetime(data['Date'])
print("\nMissing values in each column:")
print(data.isnull().sum())
data.dropna(inplace=True)
print("\nBasic statistics of the dataset:")
print(data.describe())
data['Rate Change'] = data['Unemployment Rate'].diff()
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Unemployment Rate'], label='Unemployment Rate', color='blue')
plt.axvline(pd.Timestamp('2020-03-01'), color='red', linestyle='--', label='COVID-19 Start')
plt.axvline(pd.Timestamp('2021-03-01'), color='green', linestyle='--', label='Recovery Phase Start')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid()
plt.show()
```
